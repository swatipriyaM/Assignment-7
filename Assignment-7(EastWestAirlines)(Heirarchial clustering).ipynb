```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_excel("EastWestAirlines.xlsx",sheet_name='data')
```

```python
dataset2=dataset.drop(['ID#'],axis=1)
```

```python
dataset2
```

```python
from sklearn.preprocessing import normalize
```

```python
# Normalize heterogenous numerical data 
dataset2_norm=pd.DataFrame(normalize(dataset2),columns=dataset2.columns)
dataset2_norm
```

```python
import scipy.cluster.hierarchy as sch
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(dataset2_norm,'complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clusters=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
clusters
```

```python
y=pd.DataFrame(clusters.fit_predict(dataset2_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
dataset2['clustersid']=clusters.labels_
dataset2
```

```python
dataset2.groupby('clustersid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(dataset2['clustersid'],dataset2['Balance'], c=clusters.labels_)
```
